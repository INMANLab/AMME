```python
import numpy as np
from mne.filter import notch_filter
import matplotlib.pyplot as plt

# Generate noisy signal
Fs = 1000  # Sampling frequency
t = np.arange(0, 10, 1/Fs)  # Time vector
x = np.random.randn(len(t)) + np.sin(2 * np.pi * 50 * t)  # Add 50 Hz noise

# Remove line noise
x_clean = notch_filter(x, Fs, freqs=50)

# Plot the results
plt.figure(figsize=(10, 5))
plt.subplot(2, 1, 1)
plt.plot(t, x, label="Noisy Signal")
plt.title("Original Signal with 50 Hz Noise")
plt.subplot(2, 1, 2)
plt.plot(t, x_clean, label="Cleaned Signal", color='g')
plt.title("Signal After Notch Filtering")
plt.tight_layout()
plt.show()
```
